<h1>В данной работе вычисляются коэффициенты корреляции (Пирсона, Спирмена, Кендалла), также был рассмотрен метод частной корреляции. Данный код является альтернативой некоторой части функционала IBM SPSS Statistics.

```python
import numpy as np
import pandas as pd
import pyreadstat as prs

#Чтение файла sav
file1, meta = prs.read_sav('C:/Users/79535/advert.sav')
file2, meta = prs.read_sav('C:/Users/79535/hourlywagedata.sav', usecols=["agerange","hourwage","yrsscale"])
file3, meta = prs.read_sav('C:/Users/79535/hourlywagedata.sav', usecols=["agerange","hourwage"])

print('Файл №1\n\n',file1,)
print('\nКоэффициент корреляции по методу Пирсона:\n',pd.DataFrame.corr(file1, method='pearson'))
print('\nКоэффициент корреляции по методу Спирмена:\n',pd.DataFrame.corr(file1, method='spearman'))
print('\nКоэффициент корреляции по методу Кендалла:\n',pd.DataFrame.corr(file1, method='kendall'))
print('\nКоличество строк:',len(file1.index))

print('\n\n____________________________________________\nФайл №2\n\n',file2)
print('\nКоэффициент корреляции по методу Пирсона:\n',pd.DataFrame.corr(file2, method='pearson'))
print('\nКоэффициент корреляции по методу Спирмена:\n',pd.DataFrame.corr(file2, method='spearman'))
print('\nКоэффициент корреляции по методу Кендалла:\n',pd.DataFrame.corr(file2, method='kendall'))
print('\nКоличество строк:',len(file2.index))

print('\n\n____________________________________________\nКорреляция выбранных столбцов:\n\n',file3)
print('\nКоэффициент корреляции по методу Пирсона:\n',pd.DataFrame.corr(file3, method='pearson'))
print('\nКоэффициент корреляции по методу Спирмена:\n',pd.DataFrame.corr(file3, method='spearman'))
print('\nКоэффициент корреляции по методу Кендалла:\n',pd.DataFrame.corr(file3, method='kendall'))
print('\nКоличество строк:',len(file3.index))
```
